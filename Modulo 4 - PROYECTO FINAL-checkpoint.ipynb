```python
# Carga de Librerias 

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 


from sklearn.preprocessing import MinMaxScaler, StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse, mean_squared_log_error as msle, r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
```

```python
#1. Carga del dataset para analisis de trafico para visitar sitios WEB

data = pd.read_csv('/Users/amagosr/Desktop/daily-website-visitors.csv')

data.head()
```

```python
#2. Limpieza de Datos

# Se cambia el nombre de las columnas para facilidad de uso y se eliminan los '.' de los nombre
data.columns = data.columns.str.replace('.', '', regex = True)
data.info() 
print()
data.head(5)
```

```python
# La columna 'Día' y la columna 'Día de la semana' son similares, 
# por lo que se elimina la columna 'Día'porque el tipo de columna es 'objeto'

data = data.drop('Day', axis = 1)

# 'PageLoads', 'UniqueVisits', 'FirstTimeVisits' y 'ReturningVisits' deben cambiarse al tipo 'int' 
#  data['PageLoads'].astype('int64') tambien es necesario eliminar las ','

col = ['PageLoads', 'UniqueVisits', 'FirstTimeVisits', 'ReturningVisits']
data[col] = data[col].replace(',', '', regex = True).astype('int64')

#'Fecha' Es necesario cambiar la columna 'datetime64'

data['Date'] = data['Date'].astype('datetime64')
data['year'] = data['Date'].dt.year
data['month'] = data['Date'].dt.month
data['day'] = data['Date'].dt.day
data.drop(columns = 'Date', inplace = True)
```

```python
data.info()
```

```python
#3. Se crea el conjunto de datos para entrenamiento y prueba¶

data = data.sample(frac = 1)
data['Row'] = range(1,len(data)+1)

train = data.iloc[:int(len(data) * 0.65), :]
test = data.iloc[int(len(data) * 0.65):, :]
x_use = train.drop(columns = 'ReturningVisits', axis = 0)
y_use = train['ReturningVisits']
x_submission = test.drop(columns = 'ReturningVisits', axis = 0)
y_true = test['ReturningVisits']

print('Datos de entrenamiento:', x_use.shape)
print('Datos de entrenamiento:', y_use.shape)
print('Datos de prueba:', x_submission.shape)
print('Datos de prueba:', y_true.shape)
```

```python
#4. Funcion para obtener puntuaciones.

def get_scores (model, xtrain, xtest, ytrain, ytest):
    
    train_pred = model.predict(xtrain)
    test_pred = model.predict(xtest)
    
    r2_train_score = r2_score(ytrain, train_pred)
    r2_test_score = r2_score(ytest, test_pred)
    
    diff_score = abs(r2_train_score - r2_test_score)
    
    mae_score = mae(ytest, test_pred)
    mse_score = mse(ytest, test_pred)
    rmae_score = np.sqrt(mae_score)
    rmse_score = np.sqrt(mse_score)
    
    test_pred2 = np.where(test_pred < 0, -test_pred, test_pred)
    
    msle_score = msle(ytest, test_pred2)
    rmsle_score = np.sqrt(msle_score)
    
    data = [r2_train_score, r2_test_score, diff_score, mae_score, mse_score, rmae_score, rmse_score, msle_score, rmsle_score]
    
    names = ['r2_train', 'r2_test', 'diff', 'mae', 'mse', 'rmae', 'rmse', 'msle', 'rmsle']
    
    scores = pd.Series(data, index = names)
    
    return scores
```

```python
#5. Funcion para crear modelos.

def make_models(xtrain, xtest, ytrain, ytest):
    
    models = pd.DataFrame()
    
    model_LR = LinearRegression().fit(xtrain, ytrain)
    models['model_LR'] = round(get_scores(model_LR, xtrain, xtest, ytrain, ytest),5)
        
    for d in range(3,10):

        model_RF = RandomForestRegressor(n_estimators = 1000, max_depth = d).fit(xtrain, ytrain)
        models[f'model_RF_{d}'] = round(get_scores(model_RF, xtrain, xtest, ytrain, ytest),5)
    
    for d in range(3,10):

        model_xgb = XGBRegressor(max_depth = d, objective = 'reg:squarederror').fit(xtrain, ytrain)
        models[f'model_xgb_{d}'] = round(get_scores(model_xgb, xtrain, xtest, ytrain, ytest),5)
    
    for d in range(3,10):

        model_DT = DecisionTreeRegressor(max_depth = d).fit(xtrain, ytrain)
        models[f'model_DT_{d}'] = round(get_scores(model_DT, xtrain, xtest, ytrain, ytest),5)
        
    
    models = models.T
    
    return models
```

```python
# Funcion de ajuste y contraste 

def make_models_Poly(xtrain, xtest, ytrain, ytest):
    
    models = pd.DataFrame()
    
    for a in [0.001, 0.05, 0.01, 0.5, 0.1, 1]:

        model_lasso = Lasso(alpha = a, tol = 0.1).fit(xtrain, ytrain)
        models[f'model_lasso_{a}'] = round(get_scores(model_lasso, xtrain, xtest, ytrain, ytest),5)

    for a in [0.001, 0.05, 0.01, 0.5, 0.1, 1]:

        model_ridge = Ridge(alpha = a, tol = 0.1).fit(xtrain, ytrain)
        models[f'model_ridge_{a}'] = round(get_scores(model_ridge, xtrain, xtest, ytrain, ytest),5)

    
    models = models.T
    
    return models
```

```python
#6. Se separan los datos para hacer datos de entrenamiento y prueba.

standard = StandardScaler()
minmax = MinMaxScaler()

# Caso 1
standard_xuse = standard.fit_transform(x_use)
standard_xsubmission = standard.transform(x_submission)
print(standard_xuse.shape, standard_xsubmission.shape)
xtrain_standard,xtest_standard,ytrain_standard,ytest_standard = train_test_split(standard_xuse, y_use, test_size = 0.2, random_state = 0)

# Caso 2
minmax_xuse = minmax.fit_transform(x_use)
minmax_xsubmission = minmax.transform(x_submission)
print(minmax_xuse.shape, minmax_xsubmission.shape)
xtrain_minmax,xtest_minmax,ytrain_minmax,ytest_minmax = train_test_split(minmax_xuse, y_use, test_size = 0.2, random_state = 0)
dfX = pd.concat([x_use,x_submission], ignore_index = True, axis = 0)
standard_dfX = MinMaxScaler().fit_transform(dfX)
poly_dfX = PolynomialFeatures(degree = 2).fit_transform(standard_dfX)

# Caso 3
poly_xuse = poly_dfX[:len(x_use)]
poly_xsubmission = poly_dfX[len(x_use):]
print(poly_xuse.shape, poly_xsubmission.shape)
xtrain_poly,xtest_poly,ytrain_poly,ytest_poly = train_test_split(poly_xuse, y_use, test_size = 0.3, random_state = 0)
poly_dfX2 = PolynomialFeatures(degree = 3).fit_transform(standard_dfX)

# Caso 4
poly_xuse2 = poly_dfX2[:len(x_use)]
poly_xsubmission2 = poly_dfX2[len(x_use):]
print(poly_xuse2.shape, poly_xsubmission2.shape)
xtrain_poly2,xtest_poly2,ytrain_poly2,ytest_poly2 = train_test_split(poly_xuse2, y_use, test_size = 0.3, random_state = 0)

```

```python
#7. Definición y prueba de Modelos

#Caso 1

models_standard = make_models(xtrain_standard, xtest_standard, ytrain_standard, ytest_standard)
models_standard.sort_values('rmse')
```
